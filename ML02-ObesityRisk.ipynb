### <div style="font-family: Trebuchet MS; background-color: #F2CCB6; color: #402721; padding: 12px; line-height: 1.5;">4.2.2. | Continous Column Distribution </div>
<div style="font-family: Segoe UI; line-height: 2; color: #402221; text-align: justify">
    This section will show the distribution of numerical variables in histograms, boxplots, Q-Q plots, skewness, and kurtosis values.
    <br>
    <b>Skewness</b> is an important statistical technique that helps to determine asymmetrical behavior than of the frequency distribution, or more precisely, the lack of symmetry of tails both left and right of the frequency curve.
    <blockquote style="font-size: 12px; color: #402221;">If skewness is between -0.5 and 0.5, the distribution is approximately symmetric. If skewness is between -1 and -0.5 (negative skew) or between 0.5 and 1 (positive skew), the distribution is moderately skewed. If skewness is less than -1 (negative skew) or more than 1 (positive skew), the distribution is highly skewed.</blockquote>
    The measure of <b>kurtosis</b> is the extent to which a frequency distribution is peaked in comparison with a normal curve. It is the degree of peakedness of a distribution.
    <blockquote style="font-size: 12px; color: #402221;">If kurtosis is equal to 3, the distribution is Mesokurtic. Mesokurtic is a curve having a normal peak than the normal curve.<br>If kurtosis is more than 3, the distribution is Leptokurtic. Leptokurtic is a curve having a high peak than the normal distribution.<br>If kurtosis is less than 3, the distribution is Platykurtic. Platykurtic is a curve having a low peak than the normal curve.</blockquote>
</div>

```python
train_df['family_history_with_overweight'] = (train_df['family_history_with_overweight'] == 'yes').astype('int64')
train_df['FAVC'] = (train_df['FAVC'] == 'yes').astype('int64')
train_df['SMOKE'] = (train_df['SMOKE'] == 'yes').astype('int64')
train_df['SCC'] = (train_df['SCC'] == 'yes').astype('int64')
```

## <div style="font-family: Trebuchet MS; background-color: #F2CCB6; color: #402721; padding: 12px; line-height: 1.5;">6.2. | One-Hot Encoding </div>
<div style="font-family: Segoe UI; line-height: 2; color: #402221; text-align: justify">
    The data pre-processing will be transforming categorical variables using one-hot encoding techniques.
</div>

```python
# Dummy Variables
gender = pd.get_dummies(train_df['Gender'], prefix='Gender', dtype='int64')
caec = pd.get_dummies(train_df['CAEC'], prefix='CAEC', dtype='int64')
calc = pd.get_dummies(train_df['CALC'], prefix='CALC', dtype='int64')
mtrans = pd.get_dummies(train_df['MTRANS'], prefix='MTRANS', dtype='int64')

frames = [train_df, gender, caec, calc, mtrans]
enctrain_df = pd.concat(frames, axis=1)

enctrain_df.head()
```

## <div style="font-family: Trebuchet MS; background-color: #F2CCB6; color: #402721; padding: 12px; line-height: 1.5;">6.3. | Dropping Unnecessary Variables </div>
<div style="font-family: Segoe UI; line-height: 2; color: #402221; text-align: justify">
    The unnecessary variables will be deleted.
</div>

```python
enctrain_df = enctrain_df.drop(columns=['Gender', 'CAEC', 'CALC', 'MTRANS'])
enctrain_df.head()
```

## <div style="font-family: Trebuchet MS; background-color: #F2CCB6; color: #402721; padding: 12px; line-height: 1.5;">6.4. | Preprocess Target Column </div>
<div style="font-family: Segoe UI; line-height: 2; color: #402221; text-align: justify">
    The NObeyesdad column will be transform into integer data type.
</div>

```python
enctrain_df.NObeyesdad[enctrain_df.NObeyesdad=='Insufficient_Weight'] = 0
enctrain_df.NObeyesdad[enctrain_df.NObeyesdad=='Normal_Weight'] = 1
enctrain_df.NObeyesdad[enctrain_df.NObeyesdad=='Overweight_Level_I'] = 2
enctrain_df.NObeyesdad[enctrain_df.NObeyesdad=='Overweight_Level_II'] = 3
enctrain_df.NObeyesdad[enctrain_df.NObeyesdad=='Obesity_Type_I'] = 4
enctrain_df.NObeyesdad[enctrain_df.NObeyesdad=='Obesity_Type_II'] = 5
enctrain_df.NObeyesdad[enctrain_df.NObeyesdad=='Obesity_Type_III'] = 6
enctrain_df['NObeyesdad'] = enctrain_df['NObeyesdad'].apply(np.int64)
```

## <div style="font-family: Trebuchet MS; background-color: #F2CCB6; color: #402721; padding: 12px; line-height: 1.5;">6.5. | Adding Important Column </div>
<div style="font-family: Segoe UI; line-height: 2; color: #402221; text-align: justify">
    CALC column only have 3 unique values (Frequently, Sometimes, no) instead of 4 unique values (they are missing Always value). This section will add a column that contains 0 value for CALC_Always column.
</div>

```python
enctrain_df.info()
```

```python
enctrain_df.insert(19, "CALC_Always", 0)
enctrain_df.head()
```

## <div style="font-family: Trebuchet MS; background-color: #F2CCB6; color: #402721; padding: 12px; line-height: 1.5;">6.6. | Features Separating </div>
<div style="font-family: Segoe UI; line-height: 2; color: #402221; text-align: justify">
    The NObeyesdad column will be separated.
</div>

```python
x = enctrain_df.drop(['NObeyesdad'], axis=1)
y = enctrain_df['NObeyesdad']
```

## <div style="font-family: Trebuchet MS; background-color: #F2CCB6; color: #402721; padding: 12px; line-height: 1.5;">6.7. | Data Normalization </div>
<div style="font-family: Segoe UI; line-height: 2; color: #402221; text-align: justify">
    Data normalization will use min-max normalization.
</div>

```python
x = MinMaxScaler().fit_transform(x)
```

## <div style="font-family: Trebuchet MS; background-color: #F2CCB6; color: #402721; padding: 12px; line-height: 1.5;">6.8. | Splitting the Dataset </div>
<div style="font-family: Segoe UI; line-height: 2; color: #402221; text-align: justify">
    The dataset will be splitted into 80:20 ratio (80% training and 20% testing) for each target class.
</div>

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# <div style="font-family: Trebuchet MS; background-color: #F2CCB6; color: #402721; padding: 12px; line-height: 1.5;">7. | Model Implementation </div>
<div style="font-family: Segoe UI; line-height: 2; color: #402221; text-align: justify">
    This section will implement various ML models as mentioned in Introduction section. 
</div>

## <div style="font-family: Trebuchet MS; background-color: #F2CCB6; color: #402721; padding: 12px; line-height: 1.5;">7.1. | K-Nearest Neighbor (KNN) </div>

```python
knn_clf = KNeighborsClassifier(n_neighbors=3)
knn_clf.fit(x_train, y_train)

y_pred_knn = knn_clf.predict(x_test)
```

```python
knn_acc = accuracy_score(y_pred_knn, y_test)
print('K-Nearest Neighbor Accuracy:'+'{:.2f}'.format(knn_acc*100)+'\n')

print('Classification Report\n')
print(classification_report(y_test, y_pred_knn))

print('Performance Evaluation')
fig, ((ax1, ax2)) = plt.subplots(1, 2, figsize=(10,5))

knn_matrix = ConfusionMatrix(knn_clf, ax=ax1, cmap='YlOrRd', title='K-Nearest Neighbor Confusion Matrix')
knn_matrix.fit(x_train, y_train)
knn_matrix.score(x_test, y_test)
knn_matrix.finalize()

knn_lc = LearningCurve(knn_clf, ax=ax2, title='K-Nearest Neighbor Learning Curve')
knn_lc.fit(x_train, y_train)
knn_lc.finalize()

plt.tight_layout()
```

## <div style="font-family: Trebuchet MS; background-color: #F2CCB6; color: #402721; padding: 12px; line-height: 1.5;">7.2. | Support Vector Machine (SVM) </div>

```python
svm_clf = SVC(kernel='linear', degree=1)
svm_clf.fit(x_train, y_train)

y_pred_svm = svm_clf.predict(x_test)
```

```python
svm_acc = accuracy_score(y_pred_svm, y_test)
print('Support Vector Machine Accuracy:'+'{:.2f}'.format(svm_acc*100)+'\n')

print('Classification Report\n')
print(classification_report(y_test, y_pred_svm))

print('Performance Evaluation')
fig, ((ax1, ax2)) = plt.subplots(1, 2, figsize=(10,5))

svm_matrix = ConfusionMatrix(svm_clf, ax=ax1, cmap='YlOrRd', title='Support Vector Machine Confusion Matrix')
svm_matrix.fit(x_train, y_train)
svm_matrix.score(x_test, y_test)
svm_matrix.finalize()

svm_lc = LearningCurve(svm_clf, ax=ax2, title='Support Vector Machine Learning Curve')
svm_lc.fit(x_train, y_train)
svm_lc.finalize()

plt.tight_layout()
```

## <div style="font-family: Trebuchet MS; background-color: #F2CCB6; color: #402721; padding: 12px; line-height: 1.5;">7.3. | Decision Tree </div>

```python
dt_clf = DecisionTreeClassifier(max_depth=7, min_samples_leaf=2, random_state=1)
dt_clf.fit(x_train, y_train)

y_pred_dt = dt_clf.predict(x_test)
```

```python
dt_acc = accuracy_score(y_pred_dt, y_test)
print('Decision Tree Accuracy:'+'{:.2f}'.format(dt_acc*100)+'\n')

print('Classification Report\n')
print(classification_report(y_test, y_pred_dt))

print('Performance Evaluation')
fig, ((ax1, ax2)) = plt.subplots(1, 2, figsize=(10,5))

dt_matrix = ConfusionMatrix(dt_clf, ax=ax1, cmap='YlOrRd', title='Decision Tree Confusion Matrix')
dt_matrix.fit(x_train, y_train)
dt_matrix.score(x_test, y_test)
dt_matrix.finalize()

dt_lc = LearningCurve(dt_clf, ax=ax2, title='Decision Tree Learning Curve')
dt_lc.fit(x_train, y_train)
dt_lc.finalize()

plt.tight_layout()
```

## <div style="font-family: Trebuchet MS; background-color: #F2CCB6; color: #402721; padding: 12px; line-height: 1.5;">7.4. | Model Comparison </div>

```python
compare = pd.DataFrame({'Model': ['K-Nearest Neighbor', 'Support Vector Machine', 'Decision Tree'],
                       'Accuracy': [knn_acc*100, svm_acc*100, dt_acc*100]})
compare.sort_values(by='Accuracy', ascending=False).style.background_gradient(cmap='YlOrRd').hide().set_properties(**{'font-family': 'Segoe UI'})
```

## <div style="font-family: Trebuchet MS; background-color: #F2CCB6; color: #402721; padding: 12px; line-height: 1.5;">7.5. | Model Tuning </div>

```python
params = {
    'max_depth': [5, 10, 15, 20, 25],
    'min_samples_leaf': [20, 40, 60, 80, 100],
    'criterion': ['gini', 'entropy']
}

cv = StratifiedKFold(n_splits=5)
dt_grid = GridSearchCV(estimator=dt_clf, param_grid=params, cv=cv, n_jobs=-1, verbose=1, scoring='accuracy')

dt_grid.fit(x_train, y_train)
y_pred_dt_grid = dt_grid.predict(x_test)
```

```python
dt_grid_acc = accuracy_score(y_pred_dt_grid, y_test)
print('Decision Tree with GridSearchCV Accuracy: '+'{:.2f}'.format(dt_grid_acc*100)+'\n')

print('Classification Report\n')
print(classification_report(y_test, y_pred_dt_grid))

print('Performance Evaluation')
fig, ((ax1, ax2)) = plt.subplots(1, 2, figsize=(10,5))

dt_grid_matrix = ConfusionMatrix(dt_grid, ax=ax1, cmap='YlOrRd', title='Decision Tree with GridSearchCV Confusion Matrix')
dt_grid_matrix.fit(x_train, y_train)
dt_grid_matrix.score(x_test, y_test)
dt_grid_matrix.finalize()

dt_grid_lc = LearningCurve(dt_grid, ax=ax2, title='Decision Tree with GridSearchCV Learning Curve')
dt_grid_lc.fit(x_train, y_train)
dt_grid_lc.finalize()

plt.tight_layout()
```
